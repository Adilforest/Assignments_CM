```python
# Assignment 4 - Computational Mathematics
# Task 1: Linear Curve Fitting
import numpy as np
import matplotlib.pyplot as plt

# Data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 6, 8, 11, 15])

# Method of Least Squares to compute slope (m) and intercept (c)
n = len(x)
mean_x = np.mean(x)
mean_y = np.mean(y)
m = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x) ** 2)
c = mean_y - m * mean_x

# Display results
print("Task 1: Linear Curve Fitting")
print(f"Slope (m): {m:.4f}")
print(f"Intercept (c): {c:.4f}")

# Plot data points and fitted line
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')
fitted_y = m * x + c
plt.plot(x, fitted_y, color='red', label=f'Fitted Line: y = {m:.2f}x + {c:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Curve Fitting')
plt.legend()
plt.grid()
plt.show()
```

```python
# Task 2: Polynomial Curve Fitting (Quadratic)
from numpy.polynomial.polynomial import Polynomial

# Data points
x = np.array([0, 1, 2, 3, 4])
y = np.array([2, 3, 6, 11, 18])

# Fit a quadratic polynomial using least squares
coefficients = np.polyfit(x, y, 2)
a, b, c = coefficients

# Display results
print("\nTask 2: Polynomial Curve Fitting")
print(f"Quadratic coefficients: a = {a:.4f}, b = {b:.4f}, c = {c:.4f}")

# Plot data points and fitted curve
fitted_y = a * x**2 + b * x + c
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, fitted_y, color='red', label=f'Fitted Curve: y = {a:.2f}x^2 + {b:.2f}x + {c:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Curve Fitting (Quadratic)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Task 3: Exponential Curve Fitting
def exponential_fit(x, y):
    from scipy.optimize import curve_fit

    # Define the exponential model
    def model(x, a, b):
        return a * np.exp(b * x)

    # Fit the model to the data
    params, _ = curve_fit(model, x, y, p0=(1, 0.1))
    return params

# Data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.5, 4.7, 8.8, 16.2, 30.3])

# Fit exponential model
a, b = exponential_fit(x, y)

# Display results
print("\nTask 3: Exponential Curve Fitting")
print(f"Exponential model coefficients: a = {a:.4f}, b = {b:.4f}")

# Plot data points and fitted curve
fitted_y = a * np.exp(b * x)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, fitted_y, color='red', label=f'Fitted Curve: y = {a:.2f}e^({b:.2f}x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Curve Fitting')
plt.legend()
plt.grid()
plt.show()
```

```python
# Task 4: Laws Containing Three Constants
def model_law(x, a, b, c):
    return a + b * np.log(x) + c * x**2

def fit_law(x, y):
    from scipy.optimize import curve_fit
    params, _ = curve_fit(model_law, x, y)
    return params

# Data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.1, 3.6, 6.3, 11.5, 18.9])

# Fit the model
a, b, c = fit_law(x, y)

# Display results
print("\nTask 4: Laws Containing Three Constants")
print(f"Model coefficients: a = {a:.4f}, b = {b:.4f}, c = {c:.4f}")

# Plot data points and fitted curve
fitted_y = model_law(x, a, b, c)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, fitted_y, color='red', label=f'Fitted Curve: y = {a:.2f} + {b:.2f}ln(x) + {c:.2f}x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Laws Containing Three Constants')
plt.legend()
plt.grid()
plt.show()
```

```python
# Task 5: Forward Difference Table
def forward_difference_table(x, y):
    n = len(x)
    table = np.zeros((n, n))
    table[:, 0] = y
    for j in range(1, n):
        for i in range(n - j):
            table[i, j] = table[i + 1, j - 1] - table[i, j - 1]
    return table

# Data points
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 3, 7, 13, 21])

# Compute forward difference table
table = forward_difference_table(x, y)

# Display results
print("\nTask 5: Forward Difference Table")
print("Forward Difference Table:")
print(table)
print("\nThird-order differences:", table[0, 3:])
```

```python
# Task 6: Backward Difference Table
def backward_difference_table(x, y):
    n = len(x)
    table = np.zeros((n, n))
    table[:, 0] = y
    for j in range(1, n):
        for i in range(n - 1, j - 1, -1):
            table[i, j] = table[i, j - 1] - table[i - 1, j - 1]
    return table

# Data points
x = np.array([5, 6, 7, 8, 9])
y = np.array([1, 8, 27, 64, 125])

# Compute backward difference table
table = backward_difference_table(x, y)

# Display results
print("\nTask 6: Backward Difference Table")
print("Backward Difference Table:")
print(table)
print("\nSecond-order differences:", table[-1, 2:])
```

```python
# Task 7: Higher-Order Differences
# Data points
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 8, 27, 64, 125])

# Compute forward difference table
table = forward_difference_table(x, y)

# Display results
print("\nTask 7: Higher-Order Differences")
print("Forward Difference Table:")
print(table)
print("\nFourth forward difference:", table[0, 4])
```
